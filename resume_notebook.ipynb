Now I'll loop through all of the html files in the list and generate the following headers:
job_title,job_location,body,experience (from body),experience (from bullets),tools (from body),tools (from bullets)


```python
import re

data_header = []
data_body = []
data_experience_bullets = []
data_experience_paragraph = []
data_tools_bullets = []
data_tools_paragraph = []





for i in range(0,len(file_list)):
    file =codecs.open(file_list[i],"rb")
    file1 =file.read()
    file1 =str(file1)
    page = soup(file1)
    
    header = page.find('head')
    header = header.get_text()
    
    
    data_header.append(header)
    body = page.find('body')
    body = body.get_text()
    data_body.append(body)
    
    
    bullets = page.find_all('li')
    bullet_points = []
    for bullet in bullets:
        bullet = bullet.get_text()
        bullet_points.append(bullet)
        
    experience_bullets = []
    tools_bullets = []

    for bullet in bullet_points:
        substring = ["tools","languages"]
        counter = 0
        for word in substring:
            if word in bullet.lower():
                counter += 1
        if counter > 0:
            bullet = bullet.replace("\\n","")
            tools_bullets.append(bullet)

    for bullet in bullet_points:
        substring = "experience"
        if substring in bullet.lower():
            bullet = bullet.replace("\\n","")
            experience_bullets.append(bullet)
    data_experience_bullets.append(experience_bullets)
    data_tools_bullets.append(tools_bullets)
    
    paragraphs = page.find_all('p')

    experience_paragraph = []
    tools_paragraph = []

    for paragraph in paragraphs:
        paragraph = paragraph.get_text()
        sentences = re.split('[;.]',paragraph)
        for sentence in sentences:
            substring = 'experience'
            if substring in sentence.lower():
                sentence = sentence.strip()
                sentence = sentence.replace("\\n","")
                experience_paragraph.append(sentence)

    for paragraph in paragraphs:
        paragraph = paragraph.get_text()
        sentences = re.split('[;,]',paragraph)
        for sentence in sentences:
            counter = 0
            substring = ['tools','languages']
            for word in substring:
                if word in sentence.lower():
                    counter += 1
            if counter > 0:
                sentence = sentence.strip()
                sentence = sentence.replace("\\n","")
                tools_paragraph.append(sentence)

    data_experience_paragraph.append(experience_paragraph)
    data_tools_paragraph.append(tools_paragraph)
    

data = [data_header,data_body,data_experience_bullets,data_experience_paragraph,data_tools_bullets,data_tools_paragraph]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data).transpose()
df.columns = ['Job Title','Body','Experience_From_Bullets','Experience_From_Body','Tools_From_Bullets','Tools_From_Body'] 
```
